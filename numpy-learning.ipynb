# <p style="background-color: #E7F9FF;font-family:Arial;font-size:70%;text-align:center;color:#27377C;font-weight: bold;border-radius:1000px 50px;padding:10px">Inverse of Matrix</p>

```python
arr = np.random.randint(10,50,(3,3))
arr_inv = np.linalg.inv(arr)
print(arr)
print('-'*30)
print(arr_inv)
print('-'*30)
# To ensure it is correct: their determinant of multiplication should give 1
print(np.linalg.det(np.dot(arr_inv,arr)))
```

## Another way to check


```python
# checking by thier product
res = np.dot(arr,arr_inv)
print(res)
identity = np.eye(arr.shape[0])
print('---'*20)
print(identity)
print(np.allclose(res,identity))
```
