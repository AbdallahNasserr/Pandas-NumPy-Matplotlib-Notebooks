**fontdict**

| Property  | Description                                                                         |
|-----------|-------------------------------------------------------------------------------------|
| `family`  | Font family (e.g., 'serif', 'sans-serif', 'cursive', 'fantasy', 'monospace')        |
| `style`   | Font style (e.g., 'normal', 'italic', 'oblique')                                    |
| `weight`  | Font weight (e.g., 'normal', 'bold', 'heavy', 'light', 'ultrabold', 'ultralight')   |
| `size`    | Font size (e.g., 'small', 'medium', 'large', integer values)                        |
| `color`   | Font color (e.g., 'black', 'red', '#00FF00')                                        |
| `stretch` | Font stretch (e.g., 'normal', 'condensed', 'expanded')                              |
| `variant` | Font variant (e.g., 'normal', 'small-caps')                                         |


```python
x = [1, 2, 3, 4, 5]
y = [4, 10, 12, 16, 18]
plt.style.use('ggplot')
ax = plt.axes()
fontdict = {'family':'serif','color':'blue','size':20}
plt.title('Students Grades',fontdict=fontdict)
ax.set_xlabel('Student',fontdict=fontdict)
ax.set_ylabel('Grade',fontdict=fontdict)
plt.plot(x, y, color='blue',marker= 'o', markerfacecolor='red', markeredgecolor='black',markersize=7)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Data
days = [10, 20, 30, 40, 50]
sleeping = [7, 8, 6, 11, 7]
eating = [2, 3, 4, 3, 2]
working = [7, 8, 7, 2, 2]
playing = [8, 5, 7, 8, 13]
# sleeping = sorted(sleeping)
# eating = sorted(sleeping)
# working = sorted(working)
# playing = sorted(playing)

# Define the number of bars per group
n_groups = len(days)

# Create the index for the groups
index = np.arange(n_groups)

# Define the bar width
bar_width = 0.2

# Plotting
plt.figure(figsize=(10, 6))

# Create bars
plt.bar(index, sleeping, bar_width, label='Sleeping', color='blue')
plt.bar(index + bar_width, eating, bar_width, label='Eating', color='orange')
plt.bar(index + 2 * bar_width, working, bar_width, label='Working', color='green')
plt.bar(index + 3 * bar_width, playing, bar_width, label='Playing', color='red')

# Adding titles and labels
plt.title('Daily Activities Over Days')
plt.xlabel('Days')
plt.ylabel('Hours Spent')
plt.xticks(index + 1.5 * bar_width, days)

# Adding a legend
plt.legend()

# Display the plot
plt.show()
```
